# Problem 1
    We run the below code for about five times. From these five iterations, we observe the sales1 data (represented by a blue line) to be constant and the sales2 data (represented by the red line) to vary each time when the code is executed. The random data is generated using the random poison distribution function which has a mean of 34 and consits of 12 data points.





```r
for(i in 1:5) {
    sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
    sales2<-rpois(12,34)  # random numbers, Poisson distribution, mean at 34, 12 numbers
    par(bg="cornsilk")

    plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
    title(main="Sales by Month")

    lines(sales2, type="o", pch=22, lty=2, col="red")
    grid(nx=NA, ny=NULL)
    legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
}
```

# Problem 3
    In this task we boxplot the same sales data as seen in the Problem 2. Boxplots make comparing the measures of data much more efficient. The most obvious observation is that the box plots of both sales1 (orange) and sales2 (green) are not the same. The sales2 box plot is not that much higher than the sales1 box plot. From this we can interpret that the graph suggests that there is not much of a difference between both the data of sales1 and sales2. Using this graph we can compare the range and the distribution of the sales data. The median of both the sales data approximately lie around 40. 

```r
fn<-boxplot(sales,col=c("orange","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```
